```python
import pandas as pd
import plotly.express as px
```

```python
df = pd.read_csv('nba_bet_stat.csv', skiprows=[0,1])
df
```

<h3>Ennyi pénzt veszítettem a vesztes szelvényekkel összesen:<h3>

```python
lost = int(df.loc[(df['outcome'] == 0), ['invest']].sum())
lost
```

<h3>Ennyi pénzt nyertem a nyertes szelvényekkel összesen:<h3>

```python
won = int(df.loc[(df['outcome'] == 1), ['prize']].sum())
#levonva a befektetett pénzt
won = won - int(df.loc[(df['outcome'] == 1), ['invest']].sum())
won
```

<h3>Tehát a teljes nyereség/veszteség ezek alapján:<h3>

```python
won - lost
```
